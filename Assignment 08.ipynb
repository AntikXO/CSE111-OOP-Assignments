```python
#TASK 1
class Student :
  ID = 0
  def __init__(self,name,dep,age,cg) :
    self.name = name
    self.dep = dep
    self.age = age
    self.cg = cg
    Student.ID+=1
  def showDetails(self) :
    print(f"ID: {Student.ID}")
    print(f"Name: {self.name}")
    print(f"Department: {self.dep}")
    print(f"Age: {self.age}")
    print(f"CGPA: {self.cg}")
  @classmethod
  def from_String(cls,string) :
    a =string.split("-")
    obj = cls(a[0],a[1],int(a[2]),float(a[3]))
    return obj

s1 = Student("Samin", "CSE", 21, 3.91)
s1.showDetails()
print("-----------------------")
s2 = Student("Fahim", "ECE", 21, 3.85)
s2.showDetails()
print("-----------------------")
s3 = Student("Tahura", "EEE", 22, 3.01)
s3.showDetails()
print("-----------------------")
s4 = Student.from_String("Sumaiya-BBA-23-3.96")
s4.showDetails()


#subtask 5
print("Subtask 5")
print("The difference between class variable and instance variable are: \nWe write class name to use the class variable and object name to use instance variable.\nWith this we can save something in the class variable which can be used in the other objects in order to keep track")
#Subtask 6
print("Subtask 6")
print("The difference between class method and instance method is in the class method,\nwe can change a variable in order to use it properly in the __init__ method of a class \nand also change the class variable with the help of class method.")
```

```python
#TASK 2
class Passenger:
  count = 0
  def __init__(self,name) :
    self.name = name
    self.__weight=0
    self.__busfare = 450
    Passenger.count+=1
  def set_bag_weight(self,w) :
    self.__weight = w
    if w>=20 and w<=50 :
      self.__busfare+=50
    elif w>50 :
      self.__busfare+=100
  def printDetail(self) :
    print(f"Name:{self.name}")
    print(f"Bus Fare: {self.__busfare} taka")

print("Total Passenger:", Passenger.count)
p1 = Passenger("Jack")
p1.set_bag_weight(90)
p2 = Passenger("Carol")
p2.set_bag_weight(10)
p3 = Passenger("Mike")
p3.set_bag_weight(25)
print("=========================")
p1.printDetail()
print("=========================")
p2.printDetail()
print("=========================")
p3.printDetail()
print("=========================")
print("Total Passenger:", Passenger.count)
```

```python
#TASK 3
class Travel :
  count = 0
  def __init__(self,s,d) :
    self.__source = s
    self.__destination = d
    self.__flight = "1:00"
    Travel.count+=1
  def set_time(self,time):
    self.__flight = f"{time}:00"
  def set_source(self,l):
    self.__source = l
  def set_destination(self,l) :
    self.__destination = l
  def display_travel_info(self) :
    print(f"Source: {self.__source}")
    print(f"Destination:{self.__destination}")
    return (f"Flight Time:{self.__flight}")

print("No. of Traveller =", Travel.count)
print("=======================")
t1 = Travel("Dhaka","India")
print(t1.display_travel_info())
print("=======================")
t2 = Travel("Kuala Lampur","Dhaka")
t2.set_time(23)
print(t2.display_travel_info())
print("=======================")
t3 = Travel("Dhaka","New_Zealand")
t3.set_time(15)
t3.set_destination("Germany")
print(t3.display_travel_info())
print("=======================")
t4 = Travel("Dhaka","India")
t4.set_time(9)
t4.set_source("Malaysia")
t4.set_destination("Canada")
print(t4.display_travel_info())
print("=======================")
print("No. of Traveller =", Travel.count)
```

```python
#TASK 4
class Student :
  total = 0
  brac = 0
  other = 0
  @classmethod
  def printDetails(cls) :
    print(f"Total Student(s): {cls.total}")
    print(f"BRAC University Student(s): {cls.brac}")
    print(f"Other Institution Student(s): {cls.other}")
  def __init__(self,name,dep,ins = "BRAC University") :
    self.name = name
    self.dep = dep
    self.ins = ins
    Student.total +=1
    if self.ins == "BRAC University":
      Student.brac+=1
    else :
      Student.other += 1
  @classmethod
  def createStudent(cls,name,dep,ins="BRAC University") :
    obj = cls(name,dep,ins)
    return obj
  def individualDetail(self) :
    print(f"Name: {self.name}")
    print(f"Department: {self.dep}")
    print(f"Institution: {self.ins}")

Student.printDetails()
print('#########################')
mikasa = Student('Mikasa Ackerman', "CSE")
mikasa.individualDetail()
print('------------------------------------------')
Student.printDetails()
print('========================')
harry = Student.createStudent('Harry Potter', "Defence Against Dark Arts", "Hogwarts School")
harry.individualDetail()
print('-------------------------------------------')
Student.printDetails()
print('=========================')
levi = Student.createStudent("Levi Ackerman", "CSE")
levi.individualDetail()
print('--------------------------------------------')
Student.printDetails()
```

```python
##EXTRA

#TASK 4
class Student :
  total = 0
  brac = 0
  other = 0
  @classmethod
  def printDetails(cls) :
    print(f"Total Student(s): {cls.total}")
    print(f"BRAC University Student(s): {cls.brac}")
    print(f"Other Institution Student(s): {cls.other}")
  def __init__(self,name,dep) :
    self.name = name
    self.dep = dep
    self.ins = "BRAC University"
    Student.total +=1
    if self.ins == "BRAC University":
      Student.brac+=1
  @classmethod
  def createStudent(cls,name,dep,ins="BRAC University") :
    obj = cls(name,dep)
    if ins!= "BRAC University" :
      obj.ins = ins
      cls.brac-=1
      cls.other+=1
    return obj
  def individualDetail(self) :
    print(f"Name: {self.name}")
    print(f"Department: {self.dep}")
    print(f"Institution: {self.ins}")

Student.printDetails()
print('#########################')
mikasa = Student('Mikasa Ackerman', "CSE")
mikasa.individualDetail()
print('------------------------------------------')
Student.printDetails()
print('========================')
harry = Student.createStudent('Harry Potter', "Defence Against Dark Arts", "Hogwarts School")
harry.individualDetail()
print('-------------------------------------------')
Student.printDetails()
print('=========================')
levi = Student.createStudent("Levi Ackerman", "CSE")
levi.individualDetail()
print('--------------------------------------------')
Student.printDetails()
```
