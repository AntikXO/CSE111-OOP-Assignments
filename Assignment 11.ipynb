```python
#Task 01

class Employee:
  def __init__(self,name,id,dept,salary=30000):
    self.name=name
    self.id=id
    self.dept=dept
    self.salary=salary

  def employeeDetails(self):
    print(f"Name: {self.name}, Dept: {self.dept}")
    print(f"Employee id: {self.id}, Salary: {self.salary}")

  def workDistribution(self,dept):
    if dept=="Human Resource":
      print("Collect work distribution details from the manager.")
    else:
      print("Collect work distribution details from the HR department.")

  def increment(self):
    self.salary*=1.10


class Foreign_employee(Employee):
  def __init__(self,name,id,dept,salary=30000):
    super().__init__(name,id,dept,salary)

  def employeeDetails(self):
    super().employeeDetails()
    print("Employee Type: Foreign")

  def increment(self):
    self.salary*=1.15


class Part_time_employee(Employee):
  def __init__(self,name,id,dept,salary=15000):
    super().__init__(name,id,dept,salary)

  def employeeDetails(self):
    super().employeeDetails()
    print("Employee Type: Part Time")

  def increment(self):
    print("Sadly, there is no increment for the part time employees!!")



print("1------------------------------------------")
emp1=Employee("Nawaz Ali", 102, "Marketing")
print("2------------------------------------------")
emp1.employeeDetails()
print("3------------------------------------------")
emp1.workDistribution("Marketing")
print("4------------------------------------------")
emp1.increment()
emp1.employeeDetails()
print("5------------------------------------------")
f_emp=Foreign_employee("Nadvi", 311, "Human Resource")
f_emp.employeeDetails()
print("6------------------------------------------")
f_emp.workDistribution("Human Resource")
print("7------------------------------------------")
f_emp.increment()
f_emp.employeeDetails()
print("8------------------------------------------")
p1_emp=Part_time_employee("Asif", 210, "Sales")
p2_emp=Part_time_employee("Olive", 223, "Accounts")
print("9------------------------------------------")
p1_emp.employeeDetails()
print("10------------------------------------------")
p1_emp.workDistribution("Sales")
print("11------------------------------------------")
p2_emp.increment()
print("12------------------------------------------")
p2_emp.employeeDetails()
```

```python
#Task 02

class Exam:
    def __init__(self,marks):
        self.marks = marks
        self.time = 60

    def examSyllabus(self):
        return "Maths, English"

    def examParts(self):
        return "Part 1 - Maths\nPart 2 - English\n"

class ScienceExam(Exam):
    def __init__(self,marks,time,*syllabus):
      self.marks=marks
      self.time=time
      self.syllabus=super().examSyllabus()
      self.exam_parts=super().examParts()
      self.counter=1
      self.another_counter=2
      for i in syllabus:
        self.another_counter+=1
        self.syllabus+=", "+i
        self.exam_parts+="Part "+str(self.another_counter)+ "- "+i+"\n"

      for i in self.syllabus:
        if i==",":
          self.counter+=1

    def __str__(self):
      return f"Marks: {str(self.marks)} Time: {str(self.time)} minutes Number of Parts: {str(self.counter)}"

    def examSyllabus(self):
      return self.syllabus

    def examParts(self):
      return self.exam_parts



engineering = ScienceExam(100,90,"Physics","HigherMaths")
print(engineering)
print('----------------------------------')
print(engineering.examSyllabus())
print(engineering.examParts())
print('==================================')
architecture = ScienceExam(100,120,"Physics","HigherMaths","Drawing")
print(architecture)
print('----------------------------------')
print(architecture.examSyllabus())
print(architecture.examParts())
```

```python
#Task 03

class PokemonBasic:
  def __init__(self, name = 'Default', hp = 0, weakness = 'None', type = 'Unknown'):
    self.name = name
    self.hit_point = hp
    self.weakness = weakness
    self.type = type

  def get_type(self):
    return 'Main type: ' + self.type

  def get_move(self):
    return 'Basic move: ' + 'Quick Attack'

  def __str__(self):
    return "Name: " + self.name + ", HP: " + str(self.hit_point) + ", Weakness: " + self.weakness


class PokemonExtra(PokemonBasic):
  def __init__(self,name,hp,weakness,primary_type,secondary_type=None,other_move=None):
    super().__init__(name,hp,weakness,primary_type)
    self.secondary_type=secondary_type
    self.other_move=""
    if other_move!=None:
      for i in other_move:
        if self.other_move=="":
          self.other_move+=i
        else:
          self.other_move+=", "+i

  def get_type(self):
    if self.secondary_type!=None:
      return f"{super().get_type()}, Secondary type: {self.secondary_type}"
    else:
      return super().get_type()

  def get_move(self):
    if self.other_move!="":
      return f"{super().get_move()}\nOther Move: {self.other_move}"
    else:
      return super().get_move()



print('\n------------Basic Info:--------------')
pk = PokemonBasic()
print(pk)
print(pk.get_type())
print(pk.get_move())

print('\n------------Pokemon 1 Info:-------------')
charmander = PokemonExtra('Charmander', 39, 'Water', 'Fire')
print(charmander)
print(charmander.get_type())
print(charmander.get_move())
print('\n------------Pokemon 2 Info:-------------')
charizard = PokemonExtra('Charizard', 78, 'Water', 'Fire', 'Flying', ('Fire Spin', 'Fire Blaze'))
print(charizard)
print(charizard.get_type())
print(charizard.get_move())
```

```python
#Task 04

class Cakes:

  order_list={}
  feedback_dict={}

  @classmethod
  def give_feedbacks(cls,cake,feedback):
    if cake in cls.feedback_dict:
      cls.feedback_dict[cake].append(feedback)
    else:
      cls.feedback_dict[cake]=[feedback]
    print("Thanks for your valuable feedback!")

  @classmethod
  def show_feedbacks(cls):
    print(cls.feedback_dict)

  def __init__(self,flavor,weight):
    self.flavor=flavor+" Cake"
    self.cake=flavor+" Cake "+str(weight)+"gm"
    self.weight=weight
    self.price=(self.weight/1000)*1200
    if self.cake in self.order_list:
      self.order_list[self.cake]+=1
    else:
      self.order_list[self.cake]=1
    self.sweetness="Moderate"
    self.cream_type="Whipped Cream"

  def cake_details(self):
    print(f"Flavor: {self.flavor}, Weight: {self.weight}gm\nSweetness: {self.sweetness} sugar, Cream Type: {self.cream_type}\nPrice: {self.price} Taka")

  def add_customization(self,sweetness,cream_type):
    self.sweetness=sweetness
    self.cream_type=cream_type


class Cheese_Cakes(Cakes):

  @classmethod
  def give_feedbacks(cls,cake,feedback):
    super().give_feedbacks(cake,feedback)
    print("You will get 10% discount for your next purchase!")

  def __init__(self,flavor,weight,bake_type="baked"):
    flavor=flavor+" Cheese"
    super().__init__(flavor,weight)
    self.flavor=flavor+" Cake"
    self.price=(self.weight/1000)*1500
    self.cream_type="Cream Cheese"
    self.bake_type=bake_type

  def cake_details(self):
    print(f"Flavor: {self.flavor}, Weight: {self.weight}gm\nSweetness: {self.sweetness} sugar, Cream Type: {self.cream_type}\nCake Type: {self.bake_type}, Price: {self.price} Taka")

  def add_customization(self):
    print("Sorry! No customization available for cheese cakes.")


order_1=Cakes("Chocolate",500)
order_2=Cakes("Vanilla",800)
print("(1)**********************************")
order_1.cake_details()
print("(1.1)**********************************")
print(Cakes.order_list)
print("(2)**********************************")
order_2.add_customization("Zero","Butter Cream")
order_2.cake_details()
print("(3)**********************************")
Cakes.give_feedbacks("Chocolate Cake","Very Delicious")
Cakes.give_feedbacks("Chocolate Cake","Yummy")
print("(4)**********************************")
Cakes.show_feedbacks()
print("(5)**********************************")
ch_order1=Cheese_Cakes("Red velvet",700)
ch_order1.cake_details()
print("(6)**********************************")
ch_order1.add_customization()
print("(7)**********************************")
ch_order2=Cheese_Cakes("Blue Berry",900,"No Bake")
ch_order2.cake_details()
print("(8)**********************************")
print(Cakes.order_list)
print("(9)**********************************")
Cheese_Cakes.give_feedbacks("Red velvet Cheese Cake","Average")
Cheese_Cakes.give_feedbacks("Blue Berry Cheese Cake","Liked it")
print("(10)**********************************")
Cakes.show_feedbacks()
```
