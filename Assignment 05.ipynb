```python
#TASK 01

class Student:
    def __init__(self,name,cgpa,credits=None,dept=None):
        self.name=name
        self.cgpa=cgpa
        self.credits=credits
        self.dept=dept
        self.scholarshipstatus=""
        if self.name=="Mim":
            self.credits=9
        if self.dept==None:
            self.dept="CSE"

    def checkScholarshipEligibility(self):
        if self.cgpa>=3.5 and self.credits>10:
            if self.cgpa>=3.5 and self.cgpa<3.7:
                self.scholarshipstatus="Need-based scholarship"
                print(f"{self.name} is eligible for Need-based scholarship")
            if self.cgpa>=3.7:
                self.scholarshipstatus="Merit-based scholarship"
                print(f"{self.name} is eligible for Merit-based scholarship")
        else:
            self.scholarshipstatus="No scholarship"
            print(f"{self.name} is not eligible for scholarship")
        
    def showDetails(self):
        print(f"Name: {self.name}")
        print(f"Department: {self.dept}")
        print(f"CGPA: {self.cgpa}")
        print(f"Number of Credits: {self.credits}")
        print(f"Scholarship Status: {self.scholarshipstatus}")
        

print('--------------------------')
std1 = Student("Alif", 3.99, 12)
print('--------------------------')
std1.checkScholarshipEligibility()
print('--------------------------')
std1.showDetails()
print('--------------------------')
std2 = Student("Mim", 3.4)
std3 = Student("Henry", 3.5, 15,"BBA")
print('--------------------------')
std2.checkScholarshipEligibility()
print('--------------------------')
std3.checkScholarshipEligibility()
print('--------------------------')
std2.showDetails()
print('--------------------------')
std3.showDetails()
print('--------------------------')
std4 = Student("Bob", 4.0, 6, "CSE")
print('--------------------------')
std4.checkScholarshipEligibility()
print('--------------------------')
std4.showDetails()
```

```python
#TASK 02

menu = {'Chicken Lollipop':15,'Beef Nugget':20,'Americano':180,'Red Velvet':150,'Prawn Tempura':80,'Saute Veg':200}

class Foodie:
    def __init__(self,name):
        self.name=name
        self.items_no=0
        self.items_list=[]
        self.total_spent=0
        
    def show_orders(self):
        sentence=f"""{self.name} has {self.items_no} item(s) in the cart.
Items: {self.items_list}
Total spent: {self.total_spent}"""
        return sentence
        
    def order(self,*args):
        self.items_no+=len(args)
        for i in args:
            food,quantity=i.split("-")
            self.items_list.append(food)
            total_price=menu[food]*int(quantity)
            self.total_spent+=total_price
            print(f"Ordered: {food}, quantity- {quantity}, price (per Unit)- {menu[food]}.")
            print(f"Total price- {total_price}")
           
    def pay_tips(self,tips=None):
        if tips==None:
            print("No tips to the waiter.")
        else:
            print(f"Gives {tips}/- tips to the waiter.")
            self.total_spent+=tips
            
f1 = Foodie('Frodo')
print(f1.show_orders())
print('1----------------------')
f1.order('Chicken Lollipop-3','Beef Nugget-6','Americano-1')
print('2----------------------')
print(f1.show_orders())
print('3----------------------')
f1.order('Red Velvet-1')
print('4----------------------')
f1.pay_tips(20)
print('5----------------------')
print(f1.show_orders())
f2 = Foodie('Bilbo')
print('6----------------------')
f2.order('Prawn Tempura-6','Saute Veg-1')
print('7----------------------')
f2.pay_tips()
print('8----------------------')
print(f2.show_orders())
```

```python
#TASK 03

class Department:
    def __init__(self,dept=None,sections=None):
        self.dept=dept
        self.sections=sections
        if self.dept==None:
            self.dept="ChE Department"
        if self.sections==None:
            self.sections=5
        print(f"The {self.dept} has {self.sections} sections")
        
        self.average=0
        self.count=0
        
    def add_students(self,*number):
        count=0
        if len(number)==self.sections:
            for i in number:
                count+=i
            count=round(count/len(number),2)
            self.average+=count
            self.count=(self.average*self.sections)
            print(f"The {self.dept} has an average of {count} students in each section")
        
        else:
            print(f"The {self.dept} department doesn't have {len(number)} sections")
                  
    def merge_Department(self,*merge_depts):
        for i in merge_depts:
            print(f"{i.dept} is merged to {self.dept}")
            

        for i in merge_depts:
            self.count+=(i.average*i.sections)
        print(f"Now the {self.dept} has an average of {round(self.count/self.sections,1)} students in each section.")
        
        
d1 = Department()
print('1-----------------------------')
d2 = Department('MME Department')
print('2-----------------------------')
d3 = Department('NCE Department', 8)
print('3-----------------------------')
d1.add_students(12, 23, 12, 34, 21)
print('4-----------------------------')
d2.add_students(40, 30, 21)
print('5-----------------------------')
d3.add_students(12, 34, 41, 17, 30, 22, 32, 51)
print('6-----------------------------')
mega = Department('Engineering Department', 10)
print('7-----------------------------')
mega.add_students(21,30,40,36,10,32,27,51,45,15)
print('8-----------------------------')
print(mega.merge_Department(d1, d2))
print('9-----------------------------')
print(mega.merge_Department(d3))
```

```python
#TASK 04

class Shopidify:
    def __init__ (self,name=None):
        self.name=name
        if self.name==None:
            self.name="Guest"
            print("Welcome to Shopidify")
        else:
            print(f"Welcome {self.name} to Shopidify")
        
        self.order=[]
        self.quantity=[]
                  
    def add_to_cart(self,*orders):
        for i in orders:
            if i in self.order:
                self.quantity[self.order.index(i)]+=1
            else:
                if type(i)==list:
                    for j in i:
                        if type(j)==str:
                            self.order.append(j)
                  
                        elif type(j)==int:
                            self.quantity.append(j)
                  
                           
                else:
                    if len(orders)==1:
                        self.order.append(i)
                        self.quantity.append(1)
                    else:
                        if type(i)==str:
                            self.order.append(i)
                  
                        elif type(i)==int:
                            self.quantity.append(i)
                        
    def display_cart(self):
        print(f"Items in the cart for {self.name}:")
        for i in range(len(self.order)):
            print(f"- {self.order[i]}: {self.quantity[i]}x")
            
    def checkout(self):
        print(f"Checkout completed for {self.name}")
        
    def display_history(self):
        print(f"Purchase history for {self.name}:")
        print("Transaction 1:")
        for i in range(len(self.order)):
            print(f"- {self.order[i]}: {self.quantity[i]}x")
        
            
            
guest_account = Shopidify()
print("1xxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxx")
john_account = Shopidify("John")
print("2xxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxx")
guest_account.add_to_cart("Air Jordan", 2)
guest_account.add_to_cart("Luffy Action Figure")
guest_account.display_cart()
print("3xxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxx")
john_account.add_to_cart(["Chocolate Chip Cookies", 3,"Goku Action Figure",2,"Dumbbells-5kg",2])
john_account.display_cart()
print("4xxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxx")
guest_account.add_to_cart("Air Jordan")
guest_account.display_cart()
print("5xxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxx")
guest_account.checkout()
print("6xxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxx")
guest_account.display_history()
print("7xxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxx")
john_account.checkout()
print("8xxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxx")
john_account.display_history()
print("9xxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxx")                        
```

```python
#TASK 05

class Author :
  def __init__(self,name = None) :
    self.name = name
    self.genre = {}
    self.count = 0
  def addBook(self,book_name,genre) :
    if self.name == None :
      print("A book can not be added without author name ")
    else :
      self.count+=1
      if genre not in self.genre.keys() :
        self.genre[genre]=book_name
      else :
        self.genre[genre]+=f", {book_name}"
  def setName(self,name) :
    self.name = name
  def printDetail(self) :
    print(f"Number of Book(s): {self.count} ")
    print(f"Author Name: {self.name}")
    for key,value in self.genre.items() :
      print(f"{key}: {value}")

a1 = Author()
print("=================================")
a1.addBook("Ice", "Science Fiction")
print("=================================")
a1.setName("Anna Kavan")
a1.addBook("Ice", "Science Fiction")
a1.printDetail()
print("=================================")
a2 = Author("Humayun Ahmed")
a2.addBook("Onnobhubon", "Science Fiction")
a2.addBook("Megher Upor Bari", "Horror")
print("=================================")
a2.printDetail()
a2.addBook("Ireena", "Science Fiction")
print("=================================")
a2.printDetail()
print("=================================")
```
